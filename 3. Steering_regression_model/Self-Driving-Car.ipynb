```python
from pop import AI, Camera, Pilot, Util
import cv2
```

```python
cam = Camera(300, 300)  # 입력 크기: 300x300
TF = AI.Track_Drive(cam)
TF.load_model("JMH6.h5")  # 학습된 트랙 주행 모델 로드
```

```python
Car = Pilot.AutoCar()
Car.setObstacleDistance(0)  # 장애물 무시
speed = 30
Car.forward(speed)  # 차량 전진 시작
```

```python
while True:
    try:
        img = cam.value.copy()
        t = TF.run(img)
        print("[디버깅] 추론 결과:", t)

        if 'x' in t:
            steer = float(t['x'])  # 🔧 바로 조향값으로 사용
        else:
            steer = 0  # fallback

        steer = max(min(steer, 1), -1)  # 혹시나 범위 벗어나면 제한
        print(f"[디버깅] 조향값 (steer): {steer:.3f}")
        Car.steering = steer

        # 🔵 시각화: steer가 -1~1 기준이므로 변환
        pixel_x = int((steer + 1) / 2 * 300)
        img = cv2.circle(img, (pixel_x, 230), 6, (255, 0, 0), 2)
        Util.imshow("Track Only", img)

    except KeyboardInterrupt:
        Car.stop()
        break
```
